```python
rest_type_counts = df['rest_type'].value_counts().reset_index()
rest_type_counts.columns = ['rest_type', 'count']

# Sort the DataFrame by count in descending order
rest_type_counts_sorted = rest_type_counts.sort_values(by='count', ascending=False)

# Visualization
# Create a bar plot to visualize the count of each "rest_type."

plt.figure(figsize=(12, 6))
plt.bar(rest_type_counts_sorted['rest_type'], rest_type_counts_sorted['count'], color='green')
plt.xlabel("Rest Type")
plt.ylabel("Count")
plt.title("Count of Each Rest Type in the Dataset")
plt.xticks(rotation=90)
plt.tight_layout()

# Show the bar plot
plt.show()
```

```python
restaurant_types = df.groupby('name')['rest_type'].unique().reset_index()

# Display the mapping of 'restaurant' to 'rest_type'
print(restaurant_types)

# Visualization (optional)
# You can create a bar plot to visualize the number of restaurant types for each restaurant.
# This will show which restaurant has which types of rest_type.

# Calculate the number of restaurant types for each restaurant
restaurant_types['num_rest_types'] = restaurant_types['rest_type'].apply(len)

# Sort the DataFrame by the number of restaurant types
restaurant_types_sorted = restaurant_types.sort_values(by='num_rest_types', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(restaurant_types_sorted['name'], restaurant_types_sorted['num_rest_types'])
plt.xlabel("Restaurant")
plt.ylabel("Number of Restaurant Types")
plt.title("Number of Restaurant Types for Each Restaurant")
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```
